```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import plotly.express as px
```

# Google Play Store's Mobile Apps Data Analysis

```python
df = pd.read_csv('mobileapps.csv')
```

```python
df.dropna()  # Drop rows with missing values
```

```python
df.drop_duplicates() # Remove duplicates
```

The dataset used for this analysis is a CSV file containing the following columns: App, Category, Rating, Reviews, Size, Installs, Type, Price, Content Rating, Genres, Last Updated, Current Ver, and Android Ver. The analysis was performed using Python programming language and several data analysis and visualization libraries, including Pandas, Matplotlib, and Plotly.

### Pie Chart Showing the Number of Apps per Category

Upon conducting a thorough analysis of mobile app data, it has been observed that the "Photography" category exhibits the lowest number of apps available on the Google Play Store. Conversely, the "Education" category boasts the highest number of mobile apps. These findings suggest that there may be a higher demand among Google Play Store users for educational apps compared to photography apps. Furthermore, this data implies that there could be a potential market gap for photography apps, presenting an opportunity for developers to explore novel and innovative app ideas in this category.

```python
#number of apps per category using a pie chart

# group the data by 'Category' and count the number of apps in each category
apps_per_category = df['Category'].value_counts()

#o Create a pie chart
plt.figure(figsize=(15, 15)) #change the figure size

plt.pie(apps_per_category, labels=apps_per_category.index, startangle=90, counterclock=False, autopct='%1.1f%%')

# Set title and axis labels
plt.title('Number of Apps per Category')

plt.show()
```

### Table Showing the Max Number of Installs for Each Category

The mobile apps data reveals the varying levels of installations for different categories of mobile apps on the Google Play Store. The "COMMUNICATION" category stands out with the highest number of installs, totaling over 500 million, indicating a significant demand for business-related apps. On the other hand, categories such as "AUTO_AND_VEHICLES", "BEAUTY", and "BOOKS_AND_REFERENCE" have relatively lower installations, with 50,000+ installs each, suggesting lesser popularity among users. Interestingly, the "PHOTOGRAPHY" category, also has more than 500 million installs, despite having the least amount of mobile apps in Play Store. 

```python
#groups the apps by category then shows the max number of installs for each
grouped_by_category = df.groupby('Category')
max_installs_by_category = grouped_by_category['Installs'].max()
max_installs_by_category
```

### Table Showing the Mean Rating for Each Category 

The average ratings for free apps in most categories range from 4.0 to 4.5, indicating generally positive user satisfaction. However, the "ART_AND_DESIGN" category stands out with both free and paid apps having higher average ratings of 4.4 and 4.7 respectively, suggesting that users may be willing to pay for higher-rated apps in this category.

```python
#groups the apps by category and type then shows the mean number of ratings for each
grouped_by_category_and_type = df.groupby(['Category', 'Type'])
mean_rating_by_category_and_type = grouped_by_category_and_type['Rating'].mean()

# Round to one decimal place
mean_rating_by_category_and_type = mean_rating_by_category_and_type.round(1)

mean_rating_by_category_and_type
```

### Bar Graph Showing the Number of Installs Per Category

Upon analyzing the mobile app data, it is evident that apps in all categories have installations exceeding 500 million. However, the "Communication" category stands out with a staggering 1 billion installs, indicating its widespread popularity among users. This data suggests that communication-related apps are in high demand, potentially due to the increasing reliance on mobile communication in today's digital age. It also highlights the significant market potential for developers and businesses in the communication app space. Other categories may also have substantial installations, but the "Communication" category leads the pack with the highest number of installs among the categories analyzed.

```python
#stores the columns in variables
app_names = df['Category']
reviews = df['Installs']

# makes the bar graph
plt.bar(app_names, reviews, color = "green")

# convert 'Installs' column to string
df['Installs'] = df['Installs'].astype(str)

# remove the plus sign from 'Installs' column
df['Installs'] = df['Installs'].str.replace('+', '')

# remove commas from 'Installs' column (if present)
df['Installs'] = df['Installs'].str.replace(',', '')

# convert 'Installs' column to integer
df['Installs'] = df['Installs'].astype(int)

# customizes appearance
plt.xlabel('Category')
plt.ylabel('Number of Installs')
plt.title('Bar Graph Showing the Number of Installs Per Category')
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.show()
```

### Pie Chart Showing the Number of Paid Apps compared to Free Apps

The prevalence of free apps in the app store can be attributed to user expectations and the highly competitive nature of the app market. Many users expect apps to be available for free, as there are numerous free apps already available in the market. In order to attract a larger user base and generate revenue, developers often offer apps for free and monetize them through other means such as in-app purchases, advertisements, or subscriptions. 

```python
free_df = df[df['Price'] == 0.0] #stores free apps
paid_df = df[df['Price'] > 0.0] #stores paid apps

# create data for the pie chart
labels = ['Paid Apps', 'Free Apps']
sizes = [len(paid_df), len(free_df)]
colors = ['blue', 'green']

# create a pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Paid Apps vs. Free Apps')

# show the pie chart
plt.show()
```

### Bar Graph Showing the Number of Apps per Content Rating

As seen in the bar graph, the content rating with the most apps are rated for "Everyone". Apps that target a broad audience are more prevalent in the play store due to their potential for a larger user base and higher market demand. These apps typically cater to common needs or interests shared by a wide range of users, such as communication, education, entertainment, and lifestyle. Developers often prioritize these popular categories to maximize their app's reach and potential for success in the competitive app market.

```python
#stores the different content ratings that exist
content_ratings = df['Content Rating'].value_counts()

# Create a bar chart
plt.bar(content_ratings.index, content_ratings.values, color='green', width=0.3)
plt.title('Number of Apps per Content Rating')
plt.xlabel('Content Rating')
plt.ylabel('Number of Apps')

# Show the bar chart
plt.show()
```
