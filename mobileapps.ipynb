```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import plotly.express as px
```

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
#number of apps per category using a pie chart

# group the data by 'Category' and count the number of apps in each category
apps_per_category = df['Category'].value_counts()

#o Create a pie chart
plt.figure(figsize=(20, 15)) #change the figure size

plt.pie(apps_per_category, labels=apps_per_category.index, startangle=90, counterclock=False, autopct='%1.1f%%')

# Set title and axis labels
plt.title('Number of Apps per Category')

plt.show()
```

```python
#sort by date last updated

# convert "Last Updated" column to datetime object
df['Last Updated'] = pd.to_datetime(df['Last Updated'])

## sort the dataFrame by "Last Updated" column in descending order
df = df.sort_values(by='Last Updated', ascending=False)
df
```

```python
# convert "Last Updated" column to datetime
df['Last Updated'] = pd.to_datetime(df['Last Updated'])

# extract year from "Last Updated" column
df['Year'] = df['Last Updated'].dt.year

# group by year and get the count of apps per year
line_data = df.groupby('Year').size().reset_index(name='numapps')

# create a line graph
fig = px.line(line_data, x='Year', y='numapps', title='Number of App Updates per Year')
fig.update_yaxes(title_text='Number of Apps')

#customize the line graph
fig.update_traces(line_color='green', line=dict(width=4))
fig.update_layout(height=600)

fig.show()
```

```python
free_df = df[df['Price'] == 0.0] #stores free apps
paid_df = df[df['Price'] > 0.0] #stores paid apps

# create data for the pie chart
labels = ['Paid Apps', 'Free Apps']
sizes = [len(paid_df), len(free_df)]
colors = ['blue', 'green']

# create a pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Paid Apps vs. Free Apps')

# show the pie chart
plt.show()
```

```python
#stores the different content ratings that exist
content_ratings = df['Content Rating'].value_counts()

# Create a bar chart
plt.bar(content_ratings.index, content_ratings.values, color='green', width=0.3)
plt.title('Number of Apps per Content Rating')
plt.xlabel('Content Rating')
plt.ylabel('Number of Apps')

# Show the bar chart
plt.show()
```
